```python
import requests
import pandas as pd
```

```python
tables=pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_educational_attainment')
```

```python
tables
```

```python
df=tables[1]
```

```python
df
```

```python
df = df.drop('Population over the age of 25', axis=1)
df = df.drop('With an Advanced Degree', axis=1)
```

```python
df.columns = df.columns.droplevel(0)
```

```python
df
```

```python
df = df.loc[(df['State or Territory'] == 'New York') | (df['State or Territory'] == 'Maryland') | (df['State or Territory'] == 'Hawaii') | (df['State or Territory'] == 'Illinois') | (df['State or Territory'] == 'Mississippi') | (df['State or Territory'] == 'Alabama') | (df['State or Territory'] == 'California') | (df['State or Territory'] == 'West Virginia') | (df['State or Territory'] == 'Massachusetts')]
```

```python
df
```

```python
df = df.drop(['Population'], axis = 1)
```

```python
df.columns = ['State', 'High School Diploma or higher', 'Bachelors Degree or higher']
```

```python
df
```

```python
df.to_csv('education_by_state.csv')
```
